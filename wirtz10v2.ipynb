Import libraries

```python
from statsbombpy import sb
import pandas as pd
from mplsoccer import Pitch, Sbopen, FontManager
from matplotlib import rcParams
import matplotlib.pyplot as plt

rcParams['text.color'] = '#c7d5cc'  # set the default text color
```

Locate match detail

```python
comps = sb.competitions()
comps [comps["competition_name"]=="1. Bundesliga"]
```

```python
matches = sb.matches(competition_id=9, season_id=281)
matches
```

In this project, I use match ID = 3895232, Bayern Munich VS Bayer Leverkusen match.

```python
event = sb.events(match_id=3895232)
event
```

```python
parser = Sbopen()
df, related, freeze, tactics = parser.event(3895232)
df
```

Filter out passes from Leverkusen player

```python
df = df[df['team_name']=='Bayer Leverkusen']
df
```

```python
passes = df[df['type_name']=='Pass']
passes
```

Show essential fields only

```python
passes = passes[['id', 'minute','player_id','player_name','x','y','end_x', 'end_y','pass_recipient_id','pass_recipient_name','outcome_id','outcome_name']]
passes
```

```python
passes['outcome_name'].unique()
```

Catergorize which pass is completed pass

```python
mask_complete = passes.outcome_name.isnull()
```

```python
passes_10 = passes[passes["player_name"]=='Florian Wirtz']
passes_10
```

```python
# Set up the pitch
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

# Plot the completed passes
pitch.arrows(passes_10[mask_complete].x, passes_10[mask_complete].y,
             passes_10[mask_complete].end_x, passes_10[mask_complete].end_y, width=2,
             headwidth=10, headlength=10, color='#ad993c', ax=ax, label='completed passes')

# Plot the other passes
pitch.arrows(passes_10[~mask_complete].x, passes_10[~mask_complete].y,
             passes_10[~mask_complete].end_x, passes_10[~mask_complete].end_y, width=2,
             headwidth=6, headlength=5, headaxislength=12,
             color='#ba4f45', ax=ax, label='other passes')

# Set up the legend
ax.legend(facecolor='#22312b', handlelength=5, edgecolor='None', fontsize=20, loc='upper left')

# Set the title
ax_title = ax.set_title('Florian Wirtz passes vs FC Bayern Munich', fontsize=30)
```
